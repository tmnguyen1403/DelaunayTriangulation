```python
import numpy as np
from scipy.spatial import Delaunay, delaunay_plot_2d
from matplotlib import pyplot as plt
from DebugUtils import *
```

```python
points = np.array([(-2,-2),(-2,10),(10,-2),(-1,-1)])
tri = Delaunay(points,incremental=True)
fig, axs = plt.subplots()
delaunay_plot_2d(tri, ax=axs)
plt.show()
```

```python
pi = np.array([(-1,5),(5,-1),(0,0),(1,1),(1,0),(0,2),(0.5,1)])

fig, axs = plt.subplots(3,3)
i = 0
j = 0
for p in pi:
    print(f"Add point: {p}")
    tri.add_points([p])
    delaunay_plot_2d(tri, ax=axs[i][j])
    j = j + 1
    if j >= 3:
        i += 1
        j = 0
plt.show()
```
